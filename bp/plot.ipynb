```
"""
MAKE PLOT
"""
import mmlibrary
import my_utils
import pandas as pd
import sys
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from sklearn.metrics import mean_squared_error, r2_score

init_notebook_mode(connected=True)

mmlibrary.parseArguments(sys.argv)
soda_test = my_utils.read_table_from_file("test.csv")
x1 = pd.DataFrame(dict(x1=soda_test['7oz'].values)).reset_index()
y1 = pd.DataFrame(dict(y1=soda_test['32oz'].values)).reset_index()
z1 = pd.DataFrame(dict(z1=soda_test['profit'].values)).reset_index()

x1 = x1[x1.columns[1]]
y1 = y1[y1.columns[1]]
z1 = z1[z1.columns[1]]

# soda_pred_row= mm.getBinaryFromResource("profit_df")
# soda_pred = pd.read_csv(io.StringIO(soda_pred_row.decode("utf-8")))
soda_pred = my_utils.read_table_from_file("results_df.csv")
z2 = pd.DataFrame(data=soda_pred)
z2 = z2[z2.columns[1]]

trace1 = go.Scatter3d(
    x=x1,
    y=y1,
    z=z1,
    mode='markers',
    name='actual',
    marker=dict(
        size=8,
        line=dict(
            color='rgba(217, 217, 217, 0.14)',
            width=1
        ),
        opacity=0.8
    )
)

trace2 = go.Scatter3d(
    x=x1,
    y=y1,
    z=z2,
    mode='markers',
    name='predicted',
    marker=dict(
        color='rgb(127, 127, 127)',
        size=4,
        symbol='circle',
        line=dict(
            color='rgb(204, 204, 204)',
            width=0.5
        ),
        opacity=0.6
    )
)

data = [trace1, trace2]

layout = go.Layout(
    margin=dict(l=0, r=0, b=0, t=0),
    scene=dict(
        xaxis=dict(title='7oz'),
        yaxis=dict(title='32oz'),
        zaxis=dict(title='Profit'),
    )
)
fig = go.Figure(data=data, layout=layout)

iplot(fig, filename='reression.html')
```
